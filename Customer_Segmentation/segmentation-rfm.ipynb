# üéØ **RFM Customer Segmentation Analysis - Executive Summary**

## **What is RFM Analysis?**
RFM Analysis is a powerful marketing technique that segments customers based on three key behaviors:
- **R**ecency: How recently did they purchase?
- **F**requency: How often do they purchase?
- **M**onetary: How much do they spend?

Think of it as a customer report card that tells you exactly who your best customers are and what to do with everyone else.

---

## **üîç Why This Matters for Business**

### **The Problem We're Solving:**
- 80% of revenue typically comes from 20% of customers
- Most companies treat all customers the same way
- Marketing budgets are wasted on wrong customer segments
- Customer churn happens because we don't act on early warning signs

### **The Solution:**
RFM analysis identifies **9 distinct customer types** and provides specific strategies for each, maximizing revenue while minimizing marketing waste.

---

## **üìä The 9 Customer Segments Explained**

### **üèÜ High-Value Segments (Focus Here First)**
1. **Champions (5-10% of customers)**
   - *Who they are*: Your absolute best customers - recent, frequent, high-spending
   - *Strategy*: VIP treatment, exclusive offers, referral programs
   - *Impact*: 30-40% of total revenue despite being smallest group

2. **Loyal Customers (10-15% of customers)**
   - *Who they are*: Consistent buyers who love your brand
   - *Strategy*: Upsell premium products, cross-sell complementary items
   - *Impact*: Easiest to grow revenue from

3. **Cannot Lose Them (2-5% of customers)**
   - *Who they are*: High spenders who haven't bought recently (DANGER!)
   - *Strategy*: Immediate intervention - call them personally
   - *Impact*: Losing them = massive revenue loss

### **üéØ Growth Opportunity Segments**
4. **Potential Loyalists (10-15% of customers)**
   - *Who they are*: Good customers who could become great
   - *Strategy*: Engagement campaigns, loyalty programs
   - *Impact*: Highest conversion potential

5. **New Customers (15-20% of customers)**
   - *Who they are*: Recent buyers, unknown potential
   - *Strategy*: Welcome series, onboarding, second purchase incentives
   - *Impact*: Future revenue depends on nurturing them now

### **‚ö†Ô∏è At-Risk Segments (Act Fast)**
6. **Need Attention (10-15% of customers)**
   - *Who they are*: Previously good customers showing decline
   - *Strategy*: Re-engagement campaigns, special offers
   - *Impact*: Early intervention prevents churn

7. **About to Sleep (5-10% of customers)**
   - *Who they are*: Customers drifting away
   - *Strategy*: "We miss you" campaigns, discounts
   - *Impact*: Last chance to save relationship

8. **At Risk (15-20% of customers)**
   - *Who they are*: Haven't bought in a long time, low value
   - *Strategy*: Win-back campaigns or let them go
   - *Impact*: Resource allocation decision point

### **üå± Developing Segments**
9. **Promising (10-15% of customers)**
   - *Who they are*: Recent buyers with potential
   - *Strategy*: Nurture with educational content, samples
   - *Impact*: Future growth depends on activation

### üßæ Invoice & StockCode Pattern Analysis
üîπ Invoice Format Insights
- Most invoice numbers are entirely numeric, indicating regular purchase transactions.
- A minority of invoices begin with the letter "C" ‚Äî these typically represent cancellations or returns.
- Recommended Action: Exclude "C"-prefixed invoices from RFM analysis, as they do not reflect actual purchases.
### üîπ StockCode Format Insights
- Most StockCode values are 5-digit numeric identifiers, representing actual product codes.
- Some codes may include an alphabet suffix (e.g., "85123A"), which are likely valid variations.
- A few codes are non-standard text entries (e.g., "POST", "DOTCOM"), indicating services, fees, or placeholders.
- Recommended Action: Focus only on valid product StockCodes and remove non-product/service entries before segmentation.


```python
# Step 1: Start with df_cleaned (already filtered for valid Customer ID)
df_valid = df_cleaned.copy()

# Step 2: Flag canceled invoices
df_valid['IsCanceled'] = df_valid['Invoice'].astype(str).str.startswith('C')

# Step 3: Identify valid StockCodes (either 5 digits or 5 digits + 1 letter)
def is_valid_stock(code):
    code = str(code)
    return code[:5].isdigit() and (len(code) == 5 or (len(code) == 6 and code[5].isalpha()))

df_valid['IsValidStockCode'] = df_valid['StockCode'].apply(is_valid_stock)

# Step 4: Final filter
df_segment_ready = df_valid[
    (~df_valid['IsCanceled']) & (df_valid['IsValidStockCode'])
].copy()

# Optional cleanup
df_segment_ready.drop(['IsCanceled', 'IsValidStockCode'], axis=1, inplace=True)
df_segment_ready.reset_index(drop=True, inplace=True)
df_segment_ready
```

# Advanced RFM Calculation

```python
# Set reference date (usually 1 day after the latest InvoiceDate)
analysis_date = df_segment_ready['InvoiceDate'].max() + pd.Timedelta(days=1)
# RFM grouped by Customer
rfm = df_segment_ready.groupby('Customer ID').agg({
    'InvoiceDate': [
        lambda x: (analysis_date - x.max()).days,   # Recency
        'nunique'                                   # Frequency
    ],
    'Transaction Value': 'sum'                      # Monetary
}).reset_index()

# Rename correctly: 3 columns after flattening
rfm.columns = ['CustomerID', 'Recency', 'Frequency', 'Monetary']
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
metrics = ['Recency', 'Frequency', 'Monetary']
colors = ['#66c2a5', '#fc8d62', '#8da0cb']

for i, metric in enumerate(metrics):
    sns.histplot(rfm[metric], bins=30, ax=axes[i], kde=True, color=colors[i])
    axes[i].set_title(f'Distribution of {metric}', fontsize=14)
    axes[i].set_xlabel(metric)
    axes[i].set_ylabel('Count')

plt.tight_layout()
plt.show()
```

## Outliers using IQR method

```python


# Compute IQR bounds
iqr_bounds = {}
for metric in ['Recency', 'Frequency', 'Monetary']:
    Q1 = rfm[metric].quantile(0.25)
    Q3 = rfm[metric].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    iqr_bounds[metric] = (lower, upper)

# Plot outlier thresholds
fig, axes = plt.subplots(1, 3, figsize=(16, 5))
colors = ['#66c2a5', '#fc8d62', '#8da0cb']

for i, metric in enumerate(['Recency', 'Frequency', 'Monetary']):
    lower, upper = iqr_bounds[metric]
    axes[i].bar(['Lower Bound', 'Upper Bound'], [lower, upper], color=colors[i])
    axes[i].set_title(f'{metric} IQR Outlier Thresholds', fontsize=14)
    axes[i].set_ylabel(metric)
    axes[i].grid(axis='y', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

## Intelligent Scoring System

```python
# Recency: lower is better ‚áí invert scores
rfm['R_score'] = pd.qcut(rfm['Recency'], 5, labels=[5, 4, 3, 2, 1]).astype(int)

# Frequency & Monetary: higher is better
rfm['F_score'] = pd.qcut(rfm['Frequency'].rank(method='first'), 5, labels=[1, 2, 3, 4, 5]).astype(int)
rfm['M_score'] = pd.qcut(rfm['Monetary'].rank(method='first'), 5, labels=[1, 2, 3, 4, 5]).astype(int)
rfm['RFM_string'] = rfm['R_score'].astype(str) + rfm['F_score'].astype(str) + rfm['M_score'].astype(str)
rfm['RFM_total'] = rfm[['R_score', 'F_score', 'M_score']].sum(axis=1)
```

```python
# sort by each metric for visual clarity
rfm_sorted = {
    'Recency': rfm.sort_values('Recency'),
    'Frequency': rfm.sort_values('Frequency'),
    'Monetary': rfm.sort_values('Monetary')
}

fig, axes = plt.subplots(1, 3, figsize=(16, 2))

metrics = ['Recency', 'Frequency', 'Monetary']
colors = ['YlGnBu', 'YlOrRd', 'Purples']  # Pick palettes that visually contrast

for i, metric in enumerate(metrics):
    sns.heatmap(
        [rfm_sorted[metric][metric].values],  # Keep as 1-row matrix
        cmap=colors[i],
        cbar=True,
        ax=axes[i],
        xticklabels=False,
        yticklabels=False
    )
    axes[i].set_title(f'{metric} Score Gradient', fontsize=12)

plt.tight_layout()
plt.show()
```

#  Strategic Customer Segmentation

## Advanced Segment Logic

* Champions (RFM: 555, 554, 544, 545): Best customers
* Loyal Customers (RFM: 543, 444, 435, 355): Regular buyers
* Potential Loyalists (RFM: 512, 511, 422, 421): Growth opportunity
* New Customers (RFM: 512, 511, 311, 411): Recent first-time buyers
*  Promising (RFM: 311, 411, 511): Need nurturing
* Need Attention (RFM: 155, 144, 214, 215): Declining engagement
* About to Sleep (RFM: 155, 254, 144): Risk of churning
* Cannot Lose Them (RFM: 144, 234, 343, 334): High-value at-risk
* At Risk (RFM: 111, 112, 121, 131): Likely to churn

```python
def assign_segment(rfm_str):
    if rfm_str in ['555', '554', '544', '545']:
        return 'Champions'
    elif rfm_str in ['543', '444', '435', '355']:
        return 'Loyal Customers'
    elif rfm_str in ['512', '511', '422', '421']:
        return 'Potential Loyalists'
    elif rfm_str in ['311', '411', '512', '511']:
        return 'New Customers'
    elif rfm_str in ['311', '411', '511']:
        return 'Promising'
    elif rfm_str in ['155', '144', '214', '215']:
        return 'Need Attention'
    elif rfm_str in ['155', '254', '144']:
        return 'About to Sleep'
    elif rfm_str in ['144', '234', '343', '334']:
        return 'Cannot Lose Them'
    elif rfm_str in ['111', '112', '121', '131']:
        return 'At Risk'
    else:
        return 'Other'
```

```python
rfm['Segment'] = rfm['RFM_string'].apply(assign_segment)
rfm.groupby('Segment')['RFM_total'].agg(['count', 'mean']).sort_values('count', ascending=False)
```

## Segment Profiling Logic

```python
# Total customers and total revenue for percentage calculations
total_customers = len(rfm)
total_revenue = rfm['Monetary'].sum()

# Segment-level profiling
segment_summary = rfm.groupby('Segment').agg(
    Customer_Count=('Segment', 'count'),
    Avg_Recency=('Recency', 'mean'),
    Avg_Frequency=('Frequency', 'mean'),
    Avg_Monetary=('Monetary', 'mean'),
    Total_Revenue=('Monetary', 'sum')
).reset_index()

# Add derived metrics
segment_summary['Customer_Percentage'] = (segment_summary['Customer_Count'] / total_customers * 100).round(2)
segment_summary['Revenue_Percentage'] = (segment_summary['Total_Revenue'] / total_revenue * 100).round(2)
segment_summary['CLV_Estimate'] = (segment_summary['Avg_Monetary'] * segment_summary['Avg_Frequency']).round(2)
segment_summary
```
